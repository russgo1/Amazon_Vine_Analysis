```python
# import Pandas
import pandas as pd
```

```python
# Load vine_table.csv to Pandas DataFrame
file_to_load = pd.read_csv("resources/vine_table.csv")
vine_df = pd.DataFrame(file_to_load)
```

```python
vine_df.head()
```

```python
# Sample dataframe to check for errors
vine_df.sample(10)
```

```python
# Create new dataframe that filters vine_df for rows where 
# total votes are greater than or equal to 20
more_votes_df = vine_df.loc[vine_df["total_votes"] >= 20]
more_votes_df.head()
```

```python
# Sample new dataframe to check for errors
more_votes_df.sample(20)
```

```python
# Filter new DataFrame and create new DataFrame to retrieve 
# all the rows where helpful_votes divided by total_votes is equal to or greater than 0.5 (50%)
helpful_df = more_votes_df.loc[(more_votes_df["helpful_votes"] / more_votes_df["total_votes"]) >= 0.5]
helpful_df.head()
```

```python
# Sample new dataframe to check for errors
helpful_df.sample(20)
```

```python
# Filter helpful_df and create a new DataFrame that retrieves all 
# rows where a review was written as part of the Vine program (paid)
helpful_vine_df = helpful_df.loc[helpful_df["vine"] == "Y"]
helpful_vine_df.head()
```

```python
# Sample new dataframe to check for errors
helpful_vine_df.sample(20)
```

```python
# New DataFrame to retrieve all the rows in helpful_df where 
# the review was not part of the Vine program (unpaid)
helpful_non_vine_df = helpful_df.loc[helpful_df["vine"] == "N"]
helpful_non_vine_df.head()
```

```python
# Sample new dataframe to check for errors
helpful_non_vine_df.sample(20)
```

### Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).

```python
# Paid Reviews:
# Find total number of reviews
paid_reviews = len(helpful_vine_df)
paid_rev_result = f'There are a total of {paid_reviews} paid reviews.'

# Find number of 5 star reviews
paid_5star_reviews = len(helpful_vine_df.loc[helpful_vine_df["star_rating"] == 5])
paid_5star_result = f'There are a total of {paid_5star_reviews} paid 5-star reviews.'

# Find percentage of 5 star reviews
paid_percent = (paid_5star_reviews / paid_reviews) * 100
paid_perc_result = f'{paid_percent:.2f}% of paid reviews were 5-star reviews.'

paid_outcome = f'{paid_rev_result}\n{paid_5star_result}\n{paid_perc_result}'

print(paid_outcome)
```

```python
# Unpaid Reviews:
# Find total number of reviews
unpaid_reviews = len(helpful_non_vine_df)
unpaid_rev_result = f'There are a total of {unpaid_reviews} unpaid reviews.'

# Find number of 5 star reviews
unpaid_5star_reviews = len(helpful_non_vine_df.loc[helpful_non_vine_df["star_rating"] == 5])
unpaid_5star_result = f'There are a total of {unpaid_5star_reviews} unpaid 5-star reviews.'

# Find percentage of 5 star reviews
unpaid_percent = (unpaid_5star_reviews / unpaid_reviews) * 100
unpaid_perc_result = f'{unpaid_percent:.2f}% of unpaid reviews were 5-star reviews.'

unpaid_outcome = f'{unpaid_rev_result}\n{unpaid_5star_result}\n{unpaid_perc_result}'

print(unpaid_outcome)
```

### Paid and Unpaid Data Breakdowns

```python
# Paid breakdown
print(paid_outcome)
print("\nPaid Review Data Sample:")
helpful_vine_df.sample(10)
```

```python
# Unaid breakdown
print(unpaid_outcome)
print("\nUnpaid Review Data Sample:")
helpful_non_vine_df.sample(10)
```
